## Loading and Inspecting the first few rows of the dataset
First things first. In the following code cell we will import our libraries, load our data set to the notebook, and print the first 5 rows of the DataFrame to understand its structure and content.

```python
#Import our libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# URL of our dataset housed in GitHub
url = 'https://raw.githubusercontent.com/UnitForDataScience/Data_Viz_Python_101_Workshop/refs/heads/main/Datasets/DistrictCostDatabase_2025.csv'

# Load the dataset into a pandas DataFrame
df = pd.read_csv(url)

# Display first five rows
display(df.head())
```

## Checking data types and non-null values

It's important to understand the data types of each column and identify if there are any missing values. This helps in data cleaning and feature engineering later on.

```python
#Show data types and any null values
df.info()
```

##Descriptive statistics for numerical columns

This will provide a quick summary of the central tendency, dispersion, and shape of the distribution of numerical data. It includes count, mean, standard deviation, minimum, maximum, and quartile values.

Note that this is for the overall dataset, you could specify a year or state to make this more relevant to your exploration

```python
# Show descriptive stats for the dataset
df.describe()
```

##Correlation Analysis

A correlation heatmap helps us identify relationships between different numerical variables. A higher absolute value indicates a stronger correlation (positive or negative).

Note that we are looking at all numerical variables here; you can filter these to use only the ones you are interested in.

```python
# Set the size of the plot we will print
plt.figure(figsize=(10, 10))

# Create a heatmap with all of our variables using Seaborn
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")

# Set the title
plt.title('Correlation Matrix of Numerical Features')

# Print the Heatmap
plt.show()
```

## A Simple Chart

Here is a simple line plot to get you going!

```python
# Create the plot figure and set its size (width, height)
plt.figure(figsize=(10, 5))

# Build a line chart using Seaborn
sns.lineplot(
    data=df,
    x='year',
    y='fundinggap',
    color="teal",
    linewidth=2
)

# Add our labels
plt.title("National Average Per-Pupil Spending Over Time", fontsize=14, fontweight='bold')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Spending per Student ($)", fontsize=12)

# Show the final plot
plt.show()
```
